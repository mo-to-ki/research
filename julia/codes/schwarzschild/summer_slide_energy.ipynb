```julia
# settings
M = 1
L = 1

plt = plot(
    size=(800, 800),
    xlims=(0, 10), ylims=(-1.2, 0.2),
)

x = 0.1:0.01:50
y = [ L^2/M^2 * 1/x^2 - 2/x for x in x]
plot!(plt, x, y, line=:solid, color="#ff0000", label=nothing, width=2)


plot!(plt, [0, 10], [-1, -1], line=:dash, color="#0000ff", label=nothing, width=2)
plot!(plt, [0, 10], [-0.45, -0.45], line=:dash, color="#0000ff", label=nothing, width=2)
plot!(plt, [0, 10], [0.1, 0.1], line=:dash, color="#0000ff", label=nothing, width=2)

plot!(plt, [0, 10], [0, 0], line=:dash, color="#000000", label=nothing)

plot(plt)
savefig("../../images/schwarzschild/02.png")
```

```julia
# settings
M = 1

plt = plot(
    size=(800, 800),
    xlims=(2, 50), ylims=(0.75, 1.1),
)

# L = sqrt(12)
L = 5
x = 0.1:0.01:50
y = [ - 2*L^2/M^2 * 1/x^3 + L^2/M^2 * 1/x^2 - 2/x + 1 for x in x]
plot!(plt, x, y, line=:solid, color="#008080", label=nothing, width=2)
r_positive = L^2 * (1 + sqrt(1 - 12 * M^2 / L^2)) / (2 * M)
r_negative = L^2 * (1 - sqrt(1 - 12 * M^2 / L^2)) / (2 * M)
scatter!(
    [r_positive, r_negative],
    [
        -2 * M * L^2 / r_positive^3 + L^2 / r_positive^2 - 2 * M / r_positive + 1,
        -2 * M * L^2 / r_negative^3 + L^2 / r_negative^2 - 2 * M / r_negative + 1
    ],
    color=:red, markerstrokewidth=0, label=nothing
)

L = 4.5
x = 0.1:0.01:50
y = [ - 2*L^2/M^2 * 1/x^3 + L^2/M^2 * 1/x^2 - 2/x + 1 for x in x]
plot!(plt, x, y, line=:solid, color="#008080", label=nothing, width=2)
r_positive = L^2 * (1 + sqrt(1 - 12 * M^2 / L^2)) / (2 * M)
r_negative = L^2 * (1 - sqrt(1 - 12 * M^2 / L^2)) / (2 * M)
scatter!(
    [r_positive, r_negative],
    [
        -2 * M * L^2 / r_positive^3 + L^2 / r_positive^2 - 2 * M / r_positive + 1,
        -2 * M * L^2 / r_negative^3 + L^2 / r_negative^2 - 2 * M / r_negative + 1
    ],
    color=:red, markerstrokewidth=0, label=nothing
)

L = 4
x = 0.1:0.01:50
y = [ - 2*L^2/M^2 * 1/x^3 + L^2/M^2 * 1/x^2 - 2/x + 1 for x in x]
plot!(plt, x, y, line=:solid, color="#008080", label=nothing, width=2)
r_positive = L^2 * (1 + sqrt(1 - 12 * M^2 / L^2)) / (2 * M)
r_negative = L^2 * (1 - sqrt(1 - 12 * M^2 / L^2)) / (2 * M)
scatter!(
    [r_positive, r_negative],
    [
        -2 * M * L^2 / r_positive^3 + L^2 / r_positive^2 - 2 * M / r_positive + 1,
        -2 * M * L^2 / r_negative^3 + L^2 / r_negative^2 - 2 * M / r_negative + 1
    ],
    color=:red, markerstrokewidth=0, label=nothing
)

L = 3.7
x = 0.1:0.01:50
y = [ - 2*L^2/M^2 * 1/x^3 + L^2/M^2 * 1/x^2 - 2/x + 1 for x in x]
plot!(plt, x, y, line=:solid, color="#008080", label=nothing, width=2)
r_positive = L^2 * (1 + sqrt(1 - 12 * M^2 / L^2)) / (2 * M)
r_negative = L^2 * (1 - sqrt(1 - 12 * M^2 / L^2)) / (2 * M)
scatter!(
    [r_positive, r_negative],
    [
        -2 * M * L^2 / r_positive^3 + L^2 / r_positive^2 - 2 * M / r_positive + 1,
        -2 * M * L^2 / r_negative^3 + L^2 / r_negative^2 - 2 * M / r_negative + 1
    ],
    color=:red, markerstrokewidth=0, label=nothing
)

L = sqrt(12)
x = 0.1:0.01:50
y = [ - 2*L^2/M^2 * 1/x^3 + L^2/M^2 * 1/x^2 - 2/x + 1 for x in x]
plot!(plt, x, y, line=:solid, color="#008080", label=nothing, width=2)
scatter!(
    [6],
    [ -2 * M * L^2 / 6^3 + L^2 / 6^2 - 2 * M / 6 + 1],
    color=:red, markerstrokewidth=0, label=nothing, width=3
)

L = 3
x = 0.1:0.01:50
y = [ - 2*L^2/M^2 * 1/x^3 + L^2/M^2 * 1/x^2 - 2/x + 1 for x in x]
plot!(plt, x, y, line=:solid, color="#008080", label=nothing, width=2)

plot!(plt, [6, 6], [-10, 10], line=:dash, color="#0000ff", label=nothing)


plot(plt)
savefig("../../images/schwarzschild/slide_GR01.png")
```

```julia
# settings
M = 1

anim = @animate for L in 6:-0.1:1
    x = 0.1:0.01:50
    y = [ L^2/M^2 * 1/x^2 - 2/x for x in x]
    plt = plot(
        size=(800, 800),
        xlims=(0, 50), ylims=(-0.5, 0.2),
    )
    plot!(plt, x, y, line=:solid, color="#008080", label=nothing)
    plot!(plt, [6, 6], [-10, 10], line=:dash, color="#0000ff", label=nothing)
end


gif(anim, "../../images/schwarzschild/animes/newton_circular.gif", fps = 10)
# savefig("../../images/schwarzschild/02.png")
```
