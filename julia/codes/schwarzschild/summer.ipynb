```julia
using Plots
plt = plot(
    xgrid = nothing,
    ygrid = nothing,
    zgrid = nothing,
    xaxis = false,
    yaxis = false,
    zaxis = false,
    size = (800, 800)
)

x = collect(-pi/3:0.1:pi/3)
y = collect(-pi/3:0.1:pi/3)
z = []

for i in 1:length(x)
    for j in 1:length(y)
        radius = x[i]^2 + y[j]^2
        if radius < (pi/2.)
            push!(z, sin(x[i]^2 + y[j]^2))
        else
            push!(z, 1.)
        end
    end
end

plot!(plt, x, y, z, st=:wireframe)
savefig(plt, "../../images/schwarzschild/gr.png")
```
