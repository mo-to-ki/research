## 条件 01

```julia
M = 1
L = 4

X_MIN = 0.1
X_LIM = 3.0
Y_MIN = 0.75
Y_LIM = 1.25

x = X_MIN:0.01:X_LIM

plt = plot(
    # size=(800, 800),
    xlims=(0, X_LIM),
    ylims=(Y_MIN, Y_LIM),
    grid=false,
)

plot!(plt, [0, 100] , [0, 0], line=:dot, color=:black, label=nothing, linewidth=0.5)

y_1 = [ -2 * M * L^2 / x^3 for x in x]
y_2 = [ L^2 / x^2 for x in x]
y_3 = [ - 2 * M / x for x in x]
y_total = y_1 .+ y_2 .+ y_3 .+ 1

# 修正点: findfirstの使用方法
index = findfirst(x_val -> x_val == 1.2, x)  # 1.2が存在する場合を想定
if index !== nothing
    tmp_y = y_total[index]
    println(tmp_y)
    plot!(plt, [0, 100] , [tmp_y, tmp_y], line=:dot, color=:black, label=nothing, linewidth=0.5)
end

index = findfirst(x_val -> x_val == 0.4, x)  # 0.4が存在する場合を想定
if index !== nothing
    tmp_y = y_total[index]
    println(tmp_y)
    hline!(plt, [tmp_y], line=:dot, color=:black, label=nothing, linewidth=0.5)
end

plot!(plt, x, y_1, line=:dash, color="#008080", label=nothing)
plot!(plt, x, y_2, line=:dash, color="#2f4f4f", label=nothing)
plot!(plt, x, y_3, line=:dash, color="#006400", label=nothing)
plot!(plt, x, y_total, line=:solid, color=:red, label=nothing, linewidth=2)

# savefig("../../images/schwarzschild/01.png")
plot(plt)
```

```julia
x = collect(-pi:0.01:pi)
y = collect(-pi:0.01:pi)
z = []
for x_element in x:
    for y_element in y:
        push!(z, sin(x_element^2 + y_element^2))
    end
end
plot(x,y,z, st=:wireframe)
```
