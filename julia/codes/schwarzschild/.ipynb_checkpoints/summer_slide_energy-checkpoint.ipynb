```julia
using Plots
```

## plot png

```julia
# settings
M = 1
L = 1

plt = plot(
    size=(800, 800),
    xlims=(0, 10), ylims=(-1.2, 0.2),
)

x = 0.1:0.01:50
y = [ L^2/M^2 * 1/x^2 - 2/x for x in x]
plot!(plt, x, y, line=:solid, color="#ff0000", label=nothing, width=2)


plot!(plt, [0, 10], [-1, -1], line=:dash, color="#0000ff", label=nothing)
plot!(plt, [0, 10], [-0.45, -0.45], line=:dash, color="#0000ff", label=nothing)
plot!(plt, [0, 10], [0.1, 0.1], line=:dash, color="#0000ff", label=nothing)

plot!(plt, [0, 10], [0, 0], line=:dash, color="#000000", label=nothing)

plot(plt)
# savefig("../../images/schwarzschild/02.png")
```

```julia
# settings
M = 1
L = 4.

plt = plot(
    size=(800, 800),
    xlims=(0, 50), ylims=(0.875, 1.05),
)

x = 0.1:0.01:50
y = [ -2L^2/M^2 * 1/x^3 + L^2/M^2 * 1/x^2 - 2/x + 1 for x in x]
plot!(plt, x, y, line=:solid, color="#ff0000", label=nothing)


plot!(plt, [0, 100], [1, 1], line=:dash, color="#0000ff", label=nothing)
plot!(plt, [0, 100], [0.962, 0.962], line=:dash, color="#0000ff", label=nothing)
plot!(plt, [0, 100], [0.926, 0.926], line=:dash, color="#0000ff", label=nothing)

plot(plt)
savefig("../../images/schwarzschild/general.png")
```

## plot gif

```julia
# settings
M = 1

anim = @animate for L in 6:-0.1:1
    x = 0.1:0.01:50
    y = [ L^2/M^2 * 1/x^2 - 2/x for x in x]
    plt = plot(
        size=(800, 800),
        xlims=(0, 50), ylims=(-0.5, 0.2),
    )
    plot!(plt, x, y, line=:solid, color="#008080", label=nothing)
    plot!(plt, [6, 6], [-10, 10], line=:dash, color="#0000ff", label=nothing)
end


gif(anim, "../../images/schwarzschild/animes/newton_circular.gif", fps = 10)
# savefig("../../images/schwarzschild/02.png")
```

```julia
# settings
M = 1

anim = @animate for L in 6:-0.1:1
    x = 0.1:0.01:50
    y = [ - 2*L^2/M^2 * 1/x^3 + L^2/M^2 * 1/x^2 - 2/x + 1 for x in x]
    plt = plot(
        size=(800, 800),
        xlims=(0, 50), ylims=(0.75, 1.1),
    )
    plot!(plt, x, y, line=:solid, color="#008080", label=nothing)
    plot!(plt, [6, 6], [-10, 10], line=:dash, color="#0000ff", label=nothing)
end


gif(anim, "../../images/schwarzschild/animes/general_circular.gif", fps = 10)
# savefig("../../images/schwarzschild/02.png")
```
